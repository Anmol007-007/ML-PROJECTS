```python
data = pd.read_csv("/kaggle/input/datasets/rohitgrewal/loan-approval-dataset/loan_approval_dataset.csv")
data.shape
```

```python
data.drop(columns = ["loan_id"],inplace=True, axis=1)
```

```python
data.head()
```

```python
data.columns = data.columns.str.strip()
print(data.columns)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler

le=LabelEncoder()
feature = ['education','self_employed','loan_status']
for char in feature:
    data[char]=le.fit_transform(data[char])
```

```python
data.info()
```

```python
X = data.drop(columns=['loan_status'])
Y = data['loan_status']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42)
```

```python
data.head()
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

model = Sequential()
model.add(Dense(10,activation = 'sigmoid',input_dim=11))
model.add(Dense(9,activation = 'sigmoid'))
model.add(Dense(8,activation = 'sigmoid'))
model.add(Dense(7,activation = 'sigmoid'))
model.add(Dense(6,activation = 'sigmoid'))
model.add(Dense(5,activation = 'sigmoid'))
model.add(Dense(4,activation = 'sigmoid'))
model.add(Dense(3,activation = 'sigmoid'))
model.add(Dense(2,activation = 'sigmoid'))
model.add(Dense(1,activation = 'sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer = 'Adam',metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=50)
```

```python
y_pred = model.predict(X_test)
y_prob = np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score

print("ANN Score is: ", accuracy_score(y_prob,y_test)*100)
```
