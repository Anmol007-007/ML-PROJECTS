```python
data = pd.read_csv("/kaggle/input/datasets/rjmanoj/credit-card-customer-churn-prediction/Churn_Modelling.csv")
data.head()
```

```python
data.drop(columns=["RowNumber",'Surname'],inplace=True, axis = 1)
```

```python
data['Gender'] = data['Gender'].map({'Male':1,'Female':0})
```

```python
data['Geography'].value_counts()
```

```python
data = pd.get_dummies(data,columns = ['Geography'],drop_first = True,dtype= int)
```

```python
X = data.drop(columns = ['Exited'])
```

```python
Y = data['Exited']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,y_train,y_test = train_test_split(X,Y, test_size=0.2, random_state = 42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(8,activation = 'sigmoid',input_dim = 12))
model.add(Dense(7,activation = 'sigmoid'))
model.add(Dense(6,activation = 'sigmoid'))
model.add(Dense(5,activation = 'sigmoid'))
model.add(Dense(4,activation = 'sigmoid'))
model.add(Dense(3,activation = 'sigmoid'))
model.add(Dense(2,activation = 'sigmoid'))
model.add(Dense(1,activation = 'sigmoid'))

```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'Adam',metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs =100)
```

```python
y_pred=model.predict(X_test)
```

```python
y_prob=np.where(y_pred>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score

print("ANN result is: ", accuracy_score(y_test,y_prob)*100)
```
